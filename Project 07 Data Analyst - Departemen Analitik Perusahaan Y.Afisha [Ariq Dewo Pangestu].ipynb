```python
# membuat kolom baru untuk mengidentifikasi yang menggunakan produk untul harian, minggu, dan bulan
visits['year']  = visits['start_ts'].astype('datetime64[Y]')
visits['month'] = visits['start_ts'].astype('datetime64[M]')
visits['week']  = visits['start_ts'].astype('datetime64[W]')
visits['date']  = visits['start_ts'].dt.date
```

#### Menghitung jumlah orang yang menggunakan produk setiap hari, minggu, dan bulan

Untuk menghitung jumlah orang yang menggunakan produk untuk setiap hari, minggu, dan bulan diperlukan id pengguna, dan karena diminta adalah orang-orang yang melakukan kunjungan setidaknya hanya sekali untuk id pengguna maka kunjungan untuk yang kedua atau lebih akan diabaikan.

```python
# Menghitung jumlah orang yang menggunakan produk setiap hari, minggu, dan bulan
dau_total = visits.groupby('date').agg({'uid': 'nunique'}).mean()
wau_total = visits.groupby('week').agg({'uid': 'nunique'}).mean()
mau_total = visits.groupby('month').agg({'uid': 'nunique'}).mean()

print(f'Total rata-rata DAU adalah:', int(dau_total), 'user.')
print(f'Total rata-rata WAU adalah:', int(wau_total), 'user.')
print(f'Total rata-rata MAU adalah:', int(mau_total), 'user.')
```

setelah dilakukan analisa maka didapatkan hasilnya:

- Untuk total rata-rata pengunjungan harian adalah 907 pengguna
- Untuk total rata-rata pengunjungan mingguan adalah 5724 pengguna
- Untuk total rata-rata pengunjungan bulanan adalah 23228 pengguna

```python
# Menghitung tingkat loyalitas untuk pengunjung kembali datang
# Menghitung untuk tingkat loyalitas mingguan
sticky_wau = dau_total / wau_total * 100
print('Tingkat loyalitas pengguna yang kembali dalam mingguan mengunjungin aplikasi sebesar:', int(sticky_wau),'%')
# Menghitung untuk tingkat loyalitas bulanan
sticky_mau = dau_total / mau_total * 100
print('Tingkat loyalitas pengguna yang kembali dalam bulanan mengunjungin aplikasi sebesar:', int(sticky_mau),'%')
```

```python
# Membuat history DAU pengguna untuk total menjadi perbulan
dau_monthly = {}

for month, df in visits.groupby('month'):
    key = month.strftime('%Y-%m')
    value = df.groupby('date').agg({'uid': 'nunique'}).mean()
    dau_monthly[key] = int(value)
```

```python
# Melihat hasilnya
dau_monthly
```

```python
# Membuat barplot untuk memvisualisasikan history DAU pengguna untuk total menjadi perbulan
x = [x.replace(' 00:00:00', '') for x in dau_monthly.keys()]
y = dau_monthly.values()

plt.figure(figsize=(15,5))
plt.bar(x,y)
plt.show()
```

Dilihat dari grafik, rata-rata total pengunjung yang jumlahnya sedikit ada pada bulan Juni 2017-Agustus 2017. Baru akan mengalami peningkatan di bulan September 2017 sampai Februari 2018 meskipun pada bulan Desember 2017-Januari 2018 mengalami penurunan. setelah bulan Februari 2018 jumlah rata-rata total pengunjung mengalami penurunan sampai bulan Mei 2018.

Karena diminta untuk menganalisa seberapa sering untuk pengunjung yang melakukan kunjungan kembali, mari pisahkan dan membuat variabel baru untuk semua pengunjung yang melakukan kunjungan pertama kalinya agar bisa mengetahui nanti seberapa sering pengunjung tersebut datang berkunjung kembali.

```python
# Membuat variabel untuk memfilter kunjungan pertama kali dan melihat hasilnya
first_visits = visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_visit_start_ts']
first_visits
```

Setelah membuat varibelnya, mari untuk membuat kolom baru untuk memisahkan hari, minggu, bulan, dan tahunnya.

```python
# Mengkonversi untuk menjadi hari, minggu, bulan, dan tahun
first_visits['first_visits_dt'] = first_visits['first_visit_start_ts'].dt.date
first_visits['first_visits_week'] = first_visits['first_visit_start_ts'].astype('datetime64[W]')
first_visits['first_visits_month'] = first_visits['first_visit_start_ts'].astype('datetime64[M]')
first_visits['first_visits_year'] = first_visits['first_visit_start_ts'].astype('datetime64[Y]')
```

```python
# Melihat hasilnya
first_visits
```

Semua sudah terlihat sempurna, saatnya untuk proses penggabungan ke variabel utama (visits) untuk digabung berdasarkan uid agar mengetahui uid yang melakukan kunjungannya selain kunjungan pertamanya.

```python
# Menggabungkan data visits dengan data kunjungan pertama kali untuk para pengguna
final_visits = pd.merge(visits, first_visits, on = 'uid')
```

```python
# Melihat hasilnya
final_visits.head()
```

Setelah penggabungan tabel selesai, saatnya untuk mengetahui waktu pada kunjungan pertama terhadap kunjungan lainnya berdasarkan jumlah hari dan jumlah bulannya.

```python
# Menghitung jumlah hari untuk kunjungan pertama terhadap kunjungan selanjutnya (dan konversi bulan)
final_visits['age_days'] = (final_visits['date'] - final_visits['first_visits_dt']).dt.days
final_visits['age_month'] = round(final_visits['age_days'] / 30,0).astype('int')
```

```python
# Melihat hasilnya
final_visits.head()
```

Setelah terlihat hasilnya mari kita buat analisa kohortnya dengan menggunakan pivot tabel dan melihat retensi kepada pengunjungan yang datang kembali setelah kungjungan pertamanya pada setiap bulannya berdasarkan bulan pertama setelah melakukan kunjungan pertamanya.

```python
# Membuat analisa kohort dengan agregrasi data/ pivot tabel
cohorts_visits = final_visits.pivot_table(index='first_visits_month',
                                  columns='age_month',
                                  values='uid',
                                  aggfunc='nunique')
```

```python
# Menampilkan hasilnya dan menghilangkan nilai NaN untuk bulan yang lebih baru
cohorts_visits.fillna('')
```

Setelah melihat datanya secara kasar dan sekilas, bahwa pada bulan 0 ke bulan 1 sebagian semua pengunjung yang datang kembali sepertinya tidak melebihi 10%, tapi apakah benar? mari coba konversi datanya menjadi persen.

```python
# Membuat tingkat retensi
retention = pd.DataFrame()
for col in cohorts_visits.columns:
    retention = pd.concat([retention, cohorts_visits[col] / cohorts_visits[0]], axis=1)
retention.columns = cohorts_visits.columns
```

```python
# Menampilkan graifk tingkat retensi
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Kohort: Tingkat Retensi Pengunjung')
sns.heatmap(
    retention, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmax=0.1, vmin=0
)
```

Setelah melihat tingkat retensi melalui analisa kohort bahwa untuk bulan pertama setelah para pengunjungan melakukan kunjungan pertamanya (dibulan 0) yang memiliki tingkat retensi terbesar adalah untuk kohort pada bulan September 2017 dengan nilai retensi sebear 7.4% dan yang tekecil adalah untuk kohort pada bulan Mei 2018 dengan nilai retensi sebesar 0.7%.

```python
print('Rata-rata tingkat retensi untuk pengguna yang balik setelah kedatangan pertama adalah {0:.0%}'.format(retention[1].mean()))
```

#### Menganalisa jumlah uang yang mereka kontribusikan (LTV)

```python
first_buy.head()
```

perlu membuat agregasi uid sebagai user yang nantinya dihitung berapa user yang melakukan pembelian pertamanya pada setiap bulan yang dijadikan bulan untuk pembelian pertama kalinya.

```python
# Membuat cohort untuk jumlah pengguna yang melakukan pembelian pertamanya
cohort_first_buy = first_buy.groupby('first_order_month').agg({'uid' : 'nunique'}).reset_index()
cohort_first_buy.rename(columns={'uid' : 'n_buyers'}, inplace=True)
```

```python
# Melihat hasilnya
cohort_first_buy
```

Setelah terlihat hasilnya, saatnya membuat tabel yang dimana melihat jumlah revenue untuk pembelian pertama dan pembelian selanjutnya dibulan lainnya dengan cara membuat variabel baru yang menggabungkan tabel orders dan first buy.

```python
# Menambahkan kolom untuk tanggal, minggu, dan bulan
orders['orders_date']  = orders['buy_ts'].dt.date
orders['orders_week']  = orders['buy_ts'].astype('datetime64[W]')
orders['orders_month'] = orders['buy_ts'].astype('datetime64[M]')

# Membuat kolom tambahan untuk informasi pertama kali datang dan pertama kali membeli
new_orders = pd.merge(orders, first_buy, how='inner', on='uid')
cohort_new_orders = new_orders.groupby(['first_order_month',
                                        'orders_month']).agg({'revenue' : 'sum'}).reset_index()
```

```python
# Melihat hasilnya
new_orders.head()
```

```python
# Melihat aggregasi pendapatan berdasarkan bulan pertama pembelian dan bulan pembelian lainnya
cohort_new_orders.head()
```

Setelah terlihat hasilnya saatnya membuat jeda pada pembelian pertama dibulan pertama dengan pembelian lainnya dibulan lainnya.

```python
# Menghitung jarak untuk pembelian pertama dengan pembelian kedua dan seterusnya
cohort_new_orders['age_month'] = (cohort_new_orders['orders_month'] - cohort_new_orders['first_order_month']).dt.days
cohort_new_orders['age_month'] = round(cohort_new_orders['age_month'] / 30,0).astype('int')
cohort_new_orders.columns = ['first_order_month', 'another_order_month', 'revenue', 'age_month']
```

```python
# Melihat hasilnya
cohort_new_orders.head()
```

setelah terlihat, saatnya menggabungkan kolom cohort_new_orders dengan cohort_first_buy untuk menlihat jumlah pembelinya.

```python
# Menggabungkan untuk tabel cohort_new_orders dengan cohort_first_buy
cohorts_final_orders = pd.merge(cohort_new_orders, cohort_first_buy, on='first_order_month')
```

```python
# Melihat hasilnya
cohorts_final_orders.head()
```

setelah terlihat hasilnya, sekarang bisa membuat kolom LTV dengan membagi jumlah revenue dengan total pembelinya.

```python
# Menghitung LifeTimeValue (LTV)
cohorts_final_orders['ltv'] = cohorts_final_orders['revenue'] / cohorts_final_orders['n_buyers']
```

```python
# Melihat kolom baru LTV
cohorts_final_orders.head()
```

Sekarang bisa membuat kohortnya dengan metode pivot table.

```python
# Membuat cohort untuk LTV nya
cohorts_final_orders_ltv = cohorts_final_orders.pivot_table(
    index   = 'first_order_month',
    columns = 'age_month',
    values  = 'ltv',
    aggfunc = 'mean'
)
#.cumsum(axis=1)
```

```python
# Melihat LTV nya
cohorts_final_orders_ltv.round(2).fillna('')
```

```python
# Melihat kohort LTV dengan heatmap
plt.figure(figsize=(13, 9))
plt.title('LTV')
sns.heatmap(cohorts_final_orders_ltv.cumsum(axis=1), annot=True, fmt='.2f', yticklabels = monthly_orders_by_cohorts.columns.to_series().dt.strftime('%Y-%m'), linewidths=1, linecolor='black')
plt.show()
```

Meskipun sudah terlihat, sebaiknya lebih enak ditampilkan dalam bar agar terlihat lebih jelas sumber iklan maan yang alokasi biaya banyak dan sedikit.

```python
# Menampilkan dalam bentuk bar untuk perbandingan
total_visits[~total_visits['cac'].isna()].plot(kind='bar', x='source_id', y='cac')
plt.title('CAC')
plt.xlabel('Marketing source')
plt.ylabel('Costs')
plt.show()
```

Kesimpulan:

Bahwa biaya iklan tertinggi adalah sumber iklan nomor 3 dan menghasilkan tingkat akusisi paling besar. Artinya bahwa sumber iklan nomor 3 tidaklah efisien dengan dana sebegitu besarnya hanya bisa mendapatkan pengunjung yang sedikit. Lebih baik dialokasikan dana untuk iklan ke sumber iklan yang lain seperti sumber iklan 9 dan 10 yang dimana dengan dana iklan yang sekitar 5000 dapat menghasilkan pengunjung lebih dari 10000.

#### Keuntungan Investasi tersebut bedasarkan ROI (Return On Invesment) / ROMI (Return On Marketing Investment)

sebelum menganalisa ROI dan ROMI perlu berbagai persiapan. Untuk yang pertama menganalisa jumlah pendapatan dari pembelian para pengunjung dari sumber iklannya.

```python
visits.head()
```

```python
orders.head()
```

```python
# Menggabungkan tabel visits dan orders
dataset = orders.merge(visits, how='left', left_on=['uid', 'orders_date'], right_on=['uid', 'date'])
dataset
```

Setelah terlihat, perlu mengecek nilai hilan dan duplikat agar hasilnya lebih bagus.

```python
# Mengecek nilai yang hilang
dataset.isna().sum()
```

```python
# Menghilangkan baris yang mempunya nilai hilang dan melihat jumlah baris setelah menghilangkannya
dataset = dataset.dropna()
dataset.shape
```

```python
# melihat niali yang duplikasi
dataset.duplicated().sum()
```

```python
# Melihat 5 data teratas
dataset.head()
```

Setelah selesai semua persiapan, maka dapat membuat pendapatan dari sumber iklannya.

```python
# Membuat tabel pendapatan dari sumber iklan
total_income = dataset.groupby('source_id').sum()['revenue'].reset_index()
total_income
```

Untuk menganalisa ROI/ROMI, maka perlu tabel sebelunya yaitu total_visits dan digabungkan.

```python
# Melihat tabel total_visits
total_visits
```

```python
# Mengabungkannya
total_visits = total_visits.merge(total_income, how='left', on='source_id')
total_visits
```

Setelah sudah selesai, sekarang melakukan analisa ROI/ROMI dengan cara membagi pendatapannya dengan jumlah biayanya.

```python
# Membuat kolom ROI
total_visits['roi'] = total_visits['revenue'] / total_visits['costs']
total_visits.sort_values(by='roi', ascending=False)
```

```python
# Membuat grafik perbandingan dengan bar
total_visits[~total_visits['roi'].isna()].plot(kind='bar', x='source_id', y='roi')
plt.title('Rate On Invesment')
plt.xlabel('Marketing source')
plt.ylabel('Costs')
plt.show()
```

Kesimpulan:

Setelah melakukan analisa terhadap ROI (Rate On Invesment), bahwa sumber iklan yang paling menguntungkan adalah sumber iklan 1, 2, dan 5 dengan tingkat 3 teratas. Hal ini mengartikan bahwa dengan dana yang  sebelumnya dan jumlah pengunjung barunya bisa mendapatkan tingkat ROI sebegitu besarnya.

Dan hal yang mengejutkan ternyata ada di sumber iklan nomor 3 yang dimana tingat ROI menjadi yang terendah. Padahal, perusahaan mengalokasikan dana terbesarnya pada sumber iklan nomor 3. Hal ini mempunyai arti bahwa mengalokasikan dana perusahaan sebagian besar ke sumber iklan nomor 3 akan bisa menyebabkan kerugian bagi keuangan perusahaan. Hal seperti ini harus dilaporkan langsung kepada tim pemasaran karena ketika dibiarkan lebih lama lagi akan benar-benar menggangu kinerja pendapatan pada perusahaan tersebut.
