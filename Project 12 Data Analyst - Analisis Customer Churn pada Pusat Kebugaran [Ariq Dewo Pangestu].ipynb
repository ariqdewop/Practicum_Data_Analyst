# Pengantar Proyek

Waralaba pusat kebugaran bernama Model Fitness sedang mengembangkan strategi keterlibatan pelanggan berdasarkan data analitik.

Salah satu masalah yang paling sering dijumpai oleh layanan pusat kebugaran dan bahkan layanan lainnya adalah perputaran pelanggan atau customer churn. Bagaimana cara Anda mengetahui bahwa pelanggan sudah berhenti menggunakan layanan Anda? Anda memang bisa menghitung tingkat churn berdasarkan jumlah orang yang menghapus akun atau tidak memperbarui kontrak langganan mereka. Namun terkadang, Anda tidak bisa mengetahui dengan pasti apakah seorang klien benar-benar telah berhenti: mereka mungkin meninggalkan Anda tanpa sepengetahuan Anda.

Indikator churn bisa berbeda dari satu bidang ke bidang lainnya. Jika seorang pengguna membeli dari toko daring secara tidak rutin tetapi teratur, Anda tidak bisa mengatakan bahwa pengguna tersebut telah meninggalkan toko Anda. Akan tetapi, jika selama dua minggu pengguna tidak membuka kanal toko daring yang diperbarui setiap hari, barulah Anda bisa khawatir: pengikut Anda mungkin sudah bosan dan memutuskan untuk meninggalkan Anda.

Bagi bisnis pusat kebugaran, pelanggan akan dianggap pergi jika mereka tidak lagi kembali selama satu bulan. Padahal, hal itu belum tentu benar. Seorang pelanggan bisa saja tidak mengunjungi pusat kebugaran selama satu bulan karena dia sedang berlibur, misalnya ke Bali, dan akan kembali mengunjungi pusat kebugaran setelah liburan usai. Namun, kasus seperti ini memang jarang sekali terjadi. Biasanya, jika seorang pelanggan memutuskan untuk bergabung, kemudian datang beberapa kali lalu menghilang, kemungkinan besar mereka tidak akan kembali lagi.
Untuk mengurangi tingkat churn, Model Fitness telah melakukan digitalisasi terhadap sejumlah profil pelanggannya. Adapun tugas Anda adalah menganalisis profil para pelanggan tersebut dan mengembangkan strategi retensi pelanggan (customer retention).
Untuk itu, Anda harus:

- Mempelajari cara memprediksi probabilitas churn (untuk bulan berikutnya) bagi setiap pelanggan
- Membuat segmentasi pengguna dengan memilih kelompok yang paling dominan, dan mendeskripsikan karakter utamanya
- Menganalisis faktor-faktor yang paling memengaruhi churn
- Menarik kesimpulan dasar dan memberikan rekomendasi terkait cara meningkatkan layanan pelanggan
    - Mengidentifikasi kelompok yang ditargetkan
    - Merekomendasikan langkah-langkah untuk mengurangi churn
    - Menjelaskan pola lain yang Anda temui terkait dengan interaksi pelanggan

Ada file yang akan dianalisa, yaitu `/datasets/gym_churn_us.csv` dengan format:

## Membangun Model untuk Memprediksi Churn Pengguna

Buat model klasifikasi biner untuk pelanggan di mana fitur targetnya adalah pengguna yang akan keluar bulan depan.

- Bagi datanya menjadi train set dan validation set menggunakan fungsi `train_test_split()`.
- Latih model pada train set dengan dua metode berikut:
    - regresi logistik
    - random forest
- Evaluasi accuracy, precision dan recall untuk kedua model menggunakan validation set. Gunakan untuk membandingkan model. Model manakah yang memberikan hasil terbaik?

Jangan lupa untuk mengindikasikan parameter `random_state` saat Anda membagi data dan menentukan algoritmenya.

Berdasarkan tabel korelasi sebelumnya, bahwa jenis kelamin dan nomor telepon tidak lah mempengaruhi tingkat keluar (`churn`) bagi para penggunanya. Jadi untuk variabel fitur `gender` dan `phone` tidak akan dimasukan.

```python
# Membuat matriks fitur di X dan matriks target di Y
X = df_gym.drop(['churn', 'gender', 'phone'], axis = 1)
y = df_gym['churn']

# Membagi data menjadi training set dan validation set dengan rasio 80:20
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)

# buat objek StandardScaler dan terapkan ke training set dan validation set
scaler = StandardScaler()

# latih scaler-nya pada training set menggunakan metode fit
scaler.fit(X_train)

# melatih scaler dan ubah matriks untuk training set
X_train_st = scaler.transform(X_train)

# terapkan standardisasi terhadap matriks fitur untuk validation set
X_val_st = scaler.transform(X_val)
```

```python
# mencari nilai rata-rata dan standar deviasi untuk data training dan data validasi
print('Rata-rata untuk data training:', X_train_st.mean())
print('Standar deviasi untuk data training:', X_train_st.std())
print()
print('Rata-rata untuk data validasi:', X_val_st.mean())
print('Standar deviasi untuk data validasi:', X_val_st.std())
```

Saat sudah dilakukan standardisasi bahwa untuk masing-masing rata-rata baik untuk data training dan validasi hampir mendekati 0, dan untuk standar deviasi sudah mencapai 1 dan mendekati 1 untuk  validasinya.

Setelah datanya sudah dibagi dan distandardisasi maka siap untuk di uji, tahap pengujiannya menggunakan `regresi logistik` dan `random forest`.

Evaluasi accuracy, precision dan recall untuk kedua model menggunakan validation set. Gunakan untuk membandingkan model. Model manakah yang memberikan hasil terbaik?

```python
# Membuat fungsi untuk metriks dengan uji regresi logistik dan random forest
# Mendefinisikan model
models = [LogisticRegression(random_state=0), RandomForestClassifier(random_state=0)]
models_names = ['Logistic Regression',
                'Random Forest']
# Membuat fungsi untuk uji regresi logistik dan random forest
def model_prediction(model, model_name, X_train, X_val, y_train, y_val):
    model.fit(X_train, y_train)
    y_proba = model.predict_proba(X_val)
    y_pred = model.predict(X_val)
    print('Metriks untuk model {}:\nAccuracy: {:.2f}\nPrecision: {:.2f}\nRecall: {:.2f}\n'
          .format(model_name, 
                  accuracy_score(y_val, y_pred),
                  precision_score(y_val, y_pred),
                  recall_score(y_val, y_pred)))
    return
# membuat for loop untuk menjalankan fungsi dengan kata kuncinya   
for i in range(0,len(models)):
    model_prediction(models[i], models_names[i], X_train, X_val, y_train, y_val)
```

berdasarkan uji `Logistic Regression` dan uji `Random Forest` bahwa model yang yang memberikan hasil terbaik yaitu dengan `Random Forest`. Hal ini terlihat dari nilai akurasi dan recall yang lebih baik dari `Logistic Refression` meskipun perbedaan nilai dari dua model itu terlihat sangat tipis dan nilai precision baik `Logistic Refression` maupun `Random Forest` memliki nilai yang sama yaitu 0.80.

## Membuat Klaster Pengguna

- Lakukan standardisasi terhadap datanya.
- Gunakan fungsi linkage() untuk membuat matriks jarak berdasarkan matriks fitur yang telah distandardisasi dan membuat grafik dendrogram. Catatan: pembuatan dendrogram mungkin membutuhkan waktu! Gunakan grafik yang dihasilkan untuk memperkirakan jumlah klaster yang dapat Anda pilih.
- Latih model pengklasteran dengan algoritme K-means dan prediksikan klaster pelanggannya. (Buat agar jumlah klasternya adalah `n=5`, sehingga akan lebih mudah untuk membandingkan hasil yang Anda dapat dengan hasil yang didapat siswa lain. Meski demikian, dalam kehidupan nyata, tidak ada yang akan memberi Anda petunjuk seperti itu. Oleh karena itu, ke depannya, Anda harus membuat keputusan berdasarkan grafik dari langkah sebelumnya.)
- Lihatlah nilai rata-rata fitur untuk semua klasternya. Apakah ada yang menarik perhatian Anda?
- Buatlah grafik distribusi fitur untuk setiap klasternya. Apakah Anda memperhatikan sesuatu?
- Hitunglah tingkat churn untuk setiap klaster (gunakan metode groupby()). Apakah klaster-klaster tersebut berbeda dalam hal tingkat churn? Klaster pelanggan manakah yang cenderung akan pergi, dan manakah yang akan tetap setia?

```python
# standarisasi data
X_sc = scaler.fit_transform(X)

linked = linkage(X_sc, method = 'ward') 
```

```python
# Menampilkan kluster variabel dengan dendrogram
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Pengklasteran hierarki untuk GYM')
plt.ylabel('Jarak antar kluster')
plt.xlabel('Objects kluster')
plt.show()
```

Setelah melihat matriks fitur dengan dendrogram untuk melihat kluster, bahwa berdasarkan warna kluster terbagi menjadi 3 untuk yang optimal.

Sekarang, mencoba untuk melatih model pengklasteran dengan algoritme K-means dan prediksikan klaster pelanggannya dengan membuat kluster menjadi 5.

```python
# tetapkan model k_means dengan 5 klaster
km = KMeans(n_clusters = 5)
labels = km.fit_predict(X_sc)

# simpan label klaster di kolom dataset df_gym
df_gym['cluster_km'] = labels
```

```python
# Meliaht rata-rata berdasarkan kolom cluster_km
df_gym.groupby(['cluster_km']).mean().T
```

Ada hal yang menarik untuk tabel ini, ketika untuk kluster 0 dan 2 mempunyai nilai churn yang sangat tinggal sebesar 51% dan 44% sebagai variabel target, untuk variabel fitur sebagian besar nilainya paling kecil diantara kluster-kluster lain. Sepertinya ada behavior sendiri untuk para pengguna yang akan mau keluar dengan semakin sedikit untuk beraktivitas di gym dan itu terlihat pada data di tabel teresbut tersebut dengan melihat rata-ratanya.

```python
# Membuat histogram dan distribusi fitur untuk setiap kluster
for feature in X.columns:
    sns.histplot(data=X, x=feature, hue=df_gym['cluster_km'], kde=True, palette='tab10')
    plt.title('Distribusi Fitur: {}'.format(str(feature)))
    plt.tight_layout()
    plt.xlabel('')
    plt.show()
```

kesimpulan:

Bahwa distribusi data menurut per-kluster tidak memiliki perbedaan jauh dengan distribusi menurut per-churn. Untuk `age` memiliki distribusi normal, untuk `avg_additional_charges_total`, `lifetime`, `avg_class_frequency_total`, dan `avg_class_frequency_current_month` memiliki distribusi kemiringan kekanan atau kemiringan positif.

```python
# menghitung tingkat churn untuk setiap klaster
churn_cluster = df_gym.groupby('cluster_km').agg({'churn' : 'mean'}).sort_values(by='churn', ascending=False)
churn_cluster
```

```python
# melihat tingkat churn dengan plotbar
churn_cluster.plot(kind='bar', figsize=(10,6))
plt.title('Tingkat churn berdasarkan kluster')
plt.xlabel('Kluster')
plt.show();
```

Kesimpulan:

Bahwa kluster 1 dan kluster 2 mempunyai tingka kecenderungan churn lebih tinggi, karena hampir 50% yang memilih keluar dan kluster 0 mempunyai potensi untuk kecenderungan churn yang lebih tinggi lagi karena 27% yang memilih untuk churn. Sedangkan untuk kluster 4 dan 3 mempunyai kecenderungan churn yang rendah, artinya pengguna untuk kluster ini mempunyai tingkat loyal yang tinggi atau sangat nyaman dengan kondisi tempat gymnya.
